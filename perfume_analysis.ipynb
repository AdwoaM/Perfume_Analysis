```python
mp_df = pd.read_csv('ebay_mens_perfume.csv')
wp_df = pd.read_csv('ebay_womens_perfume.csv')
```

```python
display(mp_df.head(10))
display(wp_df.head(10))
```

```python
display(mp_df.shape)
display(wp_df.shape)
```

```python
# average prices
avg_price_m = mp_df['price'].mean()
avg_price_w = wp_df['price'].mean()
print(f"Average price of men's perfumes: ${avg_price_m:.2f}")
print(f"Average price of women's perfumes: ${avg_price_w:.2f}")
```

```python
# total sales made
total_sales_m = mp_df['sold'].sum()
total_sales_w = wp_df['sold'].sum()
print(f"Total sales of men's perfumes: {total_sales_m}")
print(f"Total sales of women's perfumes: {total_sales_w}")
```

```python
# brand representation
men_brands = mp_df['brand'].nunique()
women_brands = wp_df['brand'].nunique()
print(men_brands)
print(women_brands)
```

```python
# Create data for visualization
comparison_data = {
    'Category': ['Average Price', 'Unique Brands'],
    "Men's Perfumes": [avg_price_m, men_brands],
    "Women's Perfumes": [avg_price_w, women_brands]
}

comparison_df = pd.DataFrame(comparison_data)

# Melt the dataframe for easier plotting with seaborn
melted_comparison_df = comparison_df.melt(id_vars='Category', var_name='Perfume Type', value_name='Value')

# Plot the average prices and unique brands
sns.set(style="whitegrid")

fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# Plot average prices and unique brands
sns.barplot(x='Category', y='Value', hue='Perfume Type', data=melted_comparison_df, ax=ax[0])
ax[0].set_title('Average Prices and Unique Brands Comparison')
ax[0].set_ylabel('Value')
ax[0].set_xlabel('Category')

# Plot total sales
total_sales_data = {
    'Perfume Type': ["Men's Perfumes", "Women's Perfumes"],
    'Total Sales': [total_sales_m, total_sales_w]
}
total_sales_df = pd.DataFrame(total_sales_data)

sns.barplot(x='Perfume Type', y='Total Sales', data=total_sales_df, ax=ax[1])
ax[1].set_title('Total Sales Comparison')
ax[1].set_ylabel('Total Sales')
ax[1].set_xlabel('Perfume Type')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# combine the two datasets
mp_df['sex'] = 'men'
wp_df['sex'] = 'women'

perfumes_df = pd.concat([mp_df, wp_df])
```

```python
# check for missing values
perfumes_df.isnull().sum()
```

```python
# data cleaning for missing values
perfumes_df = perfumes_df.fillna({
    'brand': 'Unknown',
    'type': 'Unknown',
    'available': 0,
    'availableText': 'Not available',
    'sold': 0,
    'lastUpdated': 'Unknown'
})
```

```python
# recheck for missing values
perfumes_df.isna().sum()
```

```python
# calculate the average price and total sales for brands
brand_sales = perfumes_df.groupby('brand')['sold'].sum().reset_index()
brand_avg_price = perfumes_df.groupby('brand')['price'].mean().reset_index()

```

```python
# Merge sales and average price data
brand_analysis_df = pd.merge(brand_sales, brand_avg_price, on='brand')
brand_analysis_df.columns = ['Brand', 'Total Sales', 'Average Price']
```

```python
# Sort brands by total sales in descending order
brand_analysis_df = brand_analysis_df.sort_values('Total Sales', ascending=False)
```

```python
# Display the top 10 brands by sales
top_brands = brand_analysis_df.head(10)
print("Top 10 Brands by Sales:")
print(top_brands)
```

```python
# Visualize the relationship between brand popularity and pricing with a bubble chart
plt.figure(figsize=(12, 8))
bubble_size = top_brands['Total Sales'] / top_brands['Total Sales'].max() * 1000
plt.scatter(top_brands['Total Sales'], top_brands['Average Price'], s=bubble_size, alpha=0.5, edgecolors="w", linewidth=2)
# Adding labels to each bubble
for i in range(len(top_brands)):
    plt.text(top_brands['Total Sales'].iloc[i], top_brands['Average Price'].iloc[i], top_brands['Brand'].iloc[i], fontsize=10, ha='right')

plt.xlabel('Total Sales')
plt.ylabel('Average Price')
plt.title('Top 10 Brands: Popularity vs. Pricing')
plt.grid(True)
plt.show()
```

### Location Analysis

```python
perfumes_df['Country'] = perfumes_df['itemLocation'].str.split(',').str[-1].str.strip()

# Group by country and calculate total sales
country_sales = perfumes_df.groupby('Country')['sold'].sum().reset_index()

# Sort countries by total sales in descending order
country_sales = country_sales.sort_values('sold', ascending=False)

# Create a choropleth map
fig = px.choropleth(
    country_sales,
    locations='Country',
    locationmode='country names',
    color='sold',
    hover_name='Country',
    color_continuous_scale=px.colors.sequential.Plasma,
    title='Geographical Distribution of Perfume Sales'
)

# Update layout for better visualization
fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True, projection_type='equirectangular'),
    margin={"r":0,"t":30,"l":0,"b":0}
)

fig.show()
```

### Sales Correlation && Availability

```python
# Calculate the correlation between availability and sales
correlation = perfumes_df['available'].corr(perfumes_df['sold'])
print(f"Correlation between availability and sales: {correlation:.2f}")

# Visualize the relationship between availability and sales
plt.figure(figsize=(10, 6))
# plt.scatter(perfumes_df['available'], perfumes_df['sold'])
plt.scatter(mp_df['available'], mp_df['sold'], alpha=0.5, label='Men', color='blue')
plt.scatter(wp_df['available'], wp_df['sold'], alpha=0.5, label='Women', color='red')
plt.xlabel('Availability')
plt.ylabel('Sales')
plt.title('Availability vs. Sales (Men v Women)')
plt.legend()
plt.show()
```

```python
## Perfume Preferences
# Group by perfume type and calculate total sales
type_sales = perfumes_df.groupby('type')['sold'].sum().reset_index()

# Sort perfume types by total sales in descending order
type_sales = type_sales.sort_values('sold', ascending=False)

# Display the perfume types by sales
print("Perfume Types by Sales:")
print(type_sales)
```

```python
# Analyze perfume type preferences by brand
brand_type_sales = perfumes_df.groupby(['brand', 'type'])['sold'].sum().reset_index()
brand_type_sales = brand_type_sales.sort_values(['brand', 'sold'], ascending=[True, False])
brand_type_sales
```

```python
# Display perfume type preferences by brand for the top 5 brands
top_brands = brand_type_sales['brand'].unique()[:5]
for brand in top_brands:
    print(f"Perfume Type Preferences for {brand}:")
    print(brand_type_sales[brand_type_sales['brand'] == brand])
    print()
```
